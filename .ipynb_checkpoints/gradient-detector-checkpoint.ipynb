```python
#Functions
def simulateLIF(dt,I,C,gL,EL,VT,W,ker,Nk,I_bias):
    V=EL*np.ones((np.shape(I)[0],2*np.shape(I)[1])) #initialising V
    stim=np.zeros((np.shape(I)[0],2*np.shape(I)[1])) #initialising synaptic stimulus
    Isyn=np.zeros((np.shape(I)[0],2*np.shape(I)[1])) #initialising synaptic current 
    spike_instants = np.zeros(np.shape(I)[0]) #spike instants for neurons
    for n in range(0,np.shape(I)[1]-1):
        neuron_refractory = np.where(n>spike_instants+(3/dt))
        if(np.size(neuron_refractory)>0):
            V[neuron_refractory,n+1] = V[neuron_refractory,n] + (dt/C)*(I[neuron_refractory,n]+Isyn[neuron_refractory,n] +I_bias[neuron_refractory,n] -gL*(V[neuron_refractory,n]-EL))
        else:
            V[spike_instants.astype(int),n+1] = EL #refractory period!
        check=np.where(V[:,n+1]>VT)[0]
        if np.size(check)>0:
            V[check,n]=VT #Artificial Spike
            spike_instants[int(check)] = n 
            stim[check,n+1:n+1+Nk] += ker
            Isyn[:,n+1:] = 1000*np.matmul(W,stim[:,n+1:])
            
    return V,Isyn

def simulateLIFNew(dt,I,C,gL,EL,VT,W,ker,Nk,I_bias):
    V=EL*np.ones((np.shape(I)[0],2*np.shape(I)[1])) #initialising V
    stim=np.zeros((np.shape(I)[0],2*np.shape(I)[1])) #initialising synaptic stimulus
    Isyn=np.zeros((np.shape(I)[0],2*np.shape(I)[1])) #initialising synaptic current 
    spike_instants = np.zeros(np.shape(I)[0])
    for n in range(0,np.shape(I)[1]-1):
        neuron_refractory = np.where(n>spike_instants+(3/dt))
        if(np.size(neuron_refractory)>0):
            V[neuron_refractory,n+1] = V[neuron_refractory,n] + (dt/C)*(I[neuron_refractory,n]+Isyn[neuron_refractory,n] +I_bias[neuron_refractory,n] -gL*(V[neuron_refractory,n]-EL))
        else:
            V[spike_instants.astype(int),n+1] = EL #refractory period!
        check=np.where(V[:,n+1]>VT)[0]
        if np.size(check)>0:
            V[check,n+1]=EL
            spike_instants[check] = n
            V[check,n]=VT #Artificial Spike
            stim[check,n+1:n+1+Nk] += ker
            Isyn[:,n+1:] = 1000*np.matmul(W[n],stim[:,n+1:])
    return V,Isyn

def currentGen(scale,num_pulses):
    ''''Current Generation with a varying pulse width

    Parameters
    ----------
    scale: number
        Factor to increase pulse width by
    num_pulses: number
        Number of pulses to generate
    Returns
    -------
    Iinp: 1D NumPy Array Current Matrix
    '''
    Iinp=np.zeros(1)
    ton = scale*(np.arange(num_pulses)+1)
    for n in range(0,len(ton)):
        Iinp = np.hstack([Iinp, 2700*np.ones(ton[n])]) 
        Iinp = np.hstack([Iinp, np.zeros(400)])
    return Iinp
    
def currentKernal(dT,tau1,tau2,kerlen):
    ''''xBased on Equation 3

    Parameters
    ----------
    dT: number
        Time-step (ms)
    tau1, tau2: number
        Time Constants
    kerlen: number
        Time Span of Kernal (ms)
    Returns
    -------
    ker,Nk: Current Kernal, Number of Kernals
    '''
    # Obtaining kernel
    Nk = np.int_(kerlen/dT)
    xker = np.linspace(0, kerlen,Nk , endpoint=True)
    ker =  np.array(np.exp(-xker/tau1)-np.exp(-xker/tau2))
    return ker,Nk


def plotter(x,y1,title,yaxlabel,y1label=None,y2=None,y2label=None,usedark=False):
    '''Graph Plotting

    Parameters
    ----------
    x:1D NumPy array
       NumPy array to be used as the x-axis data
       
   y1,y2:1D NumPy array
        NumPy array to be used as the y-axis data
     
       
    yaxlabel:String
        Y axis label
    
    y1label,y2label:String
        Legend Labels
    
     title:String
         Graph title
         
    '''
    plt.style.use('dark_background') if usedark else plt.style.use('default')
    plt.figure(figsize=(10,5)) #Customisable size
    plt.rcParams.update({'font.size': 15})
    plt.grid(True)
    plt.plot(x,y1,'g-', label = y1label)
    if(y2 is not None):plt.plot(x,y2,'r-', label = y2label),plt.legend()
    plt.title(title)
    plt.xlabel("Time/ms")
    plt.ylabel(yaxlabel)
    plt.show()
    
def currentInputGen(T,Ts,alpha,beta):
    """
    T: Temperature Input
    Ts: Set Temperature Point
    alpha,beta: constants
    
    returns: Current Input for Neuron 1(pA)
    """
    return (alpha + (beta*(T-Ts)))

def currentSineGen(lower_bound,upper_bound,dT,T):
    '''Sine Wave Generation

    Parameters
    ----------
    lower_bound: number
        Lower bound of amplitude
    upper_bound: number
        Upper bound of amplitude
    dT: number
        time-step
    T: number
        Total Time
    Returns
    -------
    current: 1D NumPy Array Current
    '''
    t = np.arange(0, T, dT)
    amplitude = (upper_bound - lower_bound)/2 #Amplitude range selection
    offset = lower_bound + amplitude
    current = (amplitude*np.sin(t)) + offset
    return current
```

### Gradient Detector

Operation required to detect temperature gradients: 

$f(t+dt)-f(t)$

The Neuron Ciruit below is the design choice for the gradient detector operation. The excitatory synapse between N4-N6 essentially computes $f(t+dt)$ and the connectivity $N_{4}-N_{5}-N_{6}$ essentially computes the $-f(t)$ operation. 

The network should only send out a spike when there is a gradient detected. Therefore, there has to be some sort of competition between the current that is comming in both pathways, and they should be trying  to cancel each other out, such that only when there is a change in N4-N6, a spike response is created.

$w_{56}$ should adapt in such a way that the net current that is comming to $N_6$ should be zero.

### Memoryless Synaptic Strength Adaptation
Modifying the strength of each synapse using the following rule:

$\Delta w_{56} = \frac{c}{\tau_a}\delta(t-t^{N_5})+\frac{d-w_{56}}{\tau_a}\Delta t$ 

$t^{N5}$ denotes the instants where neuron $N_5$ has spiked

#### Parameters in paper (Will have to be adjusted):

$c = 7.0614/Hz$

$\tau_a = 3s $ - Rate at which w56 adaptation occurs

$d = 1145.176 $  - Natural Tendency to push its weight to d

## Gradient Detector with fluctuating input current

```python
#Generate Current Input for gradient detection simulation
T = 600 #ms
dT = 0.1
iter=int(T/dT) #Number of timesteps
gradients = 6 #Number of gradients for simulation
gradient_period = int(iter/gradients) #dt for gradient switching

# 6 gradients for simulation:
#     1. Constant for t1
#     2. Positive for t2
#     3. Constant for t3
#     4. Negative for t4
#     5. Constant for t5
#     6. Positive for t6

#Initialising Currents 
lower_bound = 2800 #pA
upper_bound = 3100 #pA
positive_gradient = int((upper_bound-lower_bound)/(dT*gradient_period))
negative_gradient = -positive_gradient

Iinp = np.zeros(iter)
Iinp[:gradient_period] = lower_bound
Iinp[gradient_period:2*gradient_period] = (positive_gradient*np.linspace(0,100,gradient_period))+lower_bound
Iinp[2*gradient_period:3*gradient_period] = upper_bound
Iinp[3*gradient_period:4*gradient_period] = (negative_gradient*np.linspace(0,100,gradient_period))+upper_bound
Iinp[4*gradient_period:5*gradient_period] = lower_bound
Iinp[5*gradient_period:6*gradient_period] = (positive_gradient*np.linspace(0,100,gradient_period))+lower_bound

plt.plot(Iinp)
plt.ylabel("$I_{in}(pA)$")
plt.xlabel("Time (scale="+str(dT)+"ms/sample)")
plt.grid()
```

```python
#Parameters - Modify!
Numneurons = 3 #N4-N5-N6
I_bias = np.zeros((Numneurons,iter))
I_bias[1] = 3000 #Setting N5 bias to 5000pA

#Weights!
w45 = -1 #N4-N5 Weight
w46 = 6 #N5-N6 Weight
w56_initial = -2 #Initial N5-N6 Weight
w66 = -5 #N6-N6 weight for stability

#Weight Adaptation Parameters
c = 7.0614  #Hz^-1
tau_a = 4#s 
d = -5

#Simulate Neuron4-Neuron5
I1 = np.vstack((Iinp,np.zeros((Numneurons-1,np.shape(Iinp)[0]))))
W1 = np.zeros((Numneurons,Numneurons))
W1[1,0] = w45
Vmem1,Isyn1 = simulateLIF(dT,I1,C,gL,EL,VT,W1,ker,Nk,I_bias)
spike_times = dT*np.where(Vmem1[1,0:np.shape(I1)[1]]==VT)[0] #Spike times of N5
print("Spike Times of N5 are: " , spike_times)
stim = np.zeros(iter)
#Creating delta function compinent of weight adaptation rule
for i in range(len(spike_times)):
    stim[(int(spike_times[i]/dT))] = 1
plt.plot(stim)
plt.title("Spike Times of Neuron 5 & Weight Adaptation")
plt.ylabel("Spike")
plt.xlabel("Time (scale="+str(dT)+"ms/sample)")
plt.grid()
#Simulate Neuron4-Neuron5 Complete

'''
Apply memoryless weight adaptation rule whenever there is a spike in 
Neuron 5. The rule will increment the w56 by a certain amount for the 
i+1 iteration. If there is no spike in the next iteration we leave w56 as it is
'''
w56_vector = np.zeros(iter)
w56_vector[0] = w56_initial #initial weight, immaterial as w56 changes according to the spike instants
for i in range(0,len(w56_vector)-1):
    #Updating synaptic strengths
    if(stim[i]==1):
        w56_vector[i+1] = w56_vector[i] + ((c/tau_a)*(stim[i]) + ((d-w56_vector[i])/tau_a))
#         print(w56_vector[i+1],i)
    else:
        w56_vector[i+1] = w56_vector[i]
plt.plot(w56_vector)
plt.xlabel("Time (scale="+str(dT)+"ms/sample)")
plt.ylabel("w56")

#Upon weight adaptation rule completion, run the simualtion again with w56 matrix and for N4-N5-N6
Numneurons = 3 #N4-N5-N6
updated_weight_matrix = np.zeros((iter,3,3))
updated_weight_matrix[:,1,0] = w45 #initialising w45
updated_weight_matrix[:,2,0] = w46 #initialising w46
updated_weight_matrix[:,2,1] = w56_vector
updated_weight_matrix[:,2,2] = w66 #initialising w66

Vmem1,Isyn1 = simulateLIFNew(dT,I1,C,gL,EL,VT,updated_weight_matrix,ker,Nk,I_bias)#Simulate with weight adaptation

#Plotting simulation
t = dt*(np.arange(len(Iinp)))
plt.style.use('seaborn-darkgrid')
f = plt.figure(figsize=(30,15))
plt.grid(True)
plt.plot(t ,Iinp/2700 , 'b-',label = "Input Current") #Current scaled
for n in range(0,Numneurons):
    mem_label = ["Vmem4","Vmem5","Vmem6"]
#     syn_label = ["Isyn45","Isyn46","Isyn56"]
    plt.plot(t,2+n+(Vmem1[n,0:np.shape(I1)[1]]-EL)/(3*VT-EL) , 'r-',label = mem_label[n])
    plt.plot(t,5+n+(Isyn1[n,0:np.shape(I1)[1]])/(2700) , 'g-')
plt.xlabel('t (ms)')
plt.axis([0,600,0,8])
plt.legend()
plt.show(block = False)
```

```python
def simulateLIF(dT,I,C,gL,EL,VT):
    V = EL*np.ones((np.shape(I)))
    #iterating in the temporal dimension
    spike_instant = 0
    for n in (range((np.shape(I))[1]-1)):
        if(n>spike_instant+(3/dT)):
            V[:,n+1] = V[:,n] + (dT/C) * (-gL*(V[:,n]-EL)+I[:,n])
        else:
            V[:,n+1] = EL
        check=np.where(V[:,n+1]>VT)[0]
        if np.size(check)>0:
            V[check,n+1]=EL
            spike_instant = n
            V[check,n]= VT #Artificial Spike
    return V,spike_instant


def createCurrentMatrix(Ic,k,alpha,dT,T):
    '''
    We are assuming current is not fluctuating from t=1 to t=T
    
    Parameters
    ----------
    Ic - steady state current to initiate a spike
    k - number of neurons
    alpha - constant
    dT - time step
    T - total simulation time
    
    Returns
    -------
    I : k x (T/dT) - D NumPy array (pA)
    '''
    current = []
    n=1
    n_columns = int(T/dT)
    I = np.empty((k,n_columns), dtype=object)
    I[:,0] = 0 #at t=0, we are assuming the neuron is in steady state with 0 current
    for i in range(k):
        i_input = Ic*(1+n*alpha)
        current.append(i_input)
        n+=1
    for i in range(1,n_columns):
        I[:,i] = current
    return I

#Problem 1 
Ic=2700
neurons=1
alpha=0.1
C=300
gL=30
EL=-70
VT=20.0
T=500  
dt= .1 
Iapp = createCurrentMatrix(Ic,neurons,alpha,dt,T)
Vout,spike_instant=simulateLIF(dt,Iapp,C,gL,EL,VT)
plt.plot(Vout[0])
```
